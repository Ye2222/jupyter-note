```python
import pandas as pd
```

```python
df = pd.read_excel('2019会计师事务所100强榜单.xlsx')
```

```python
df
```

```python
def excel_one_line_to_list(df, col):
    df = df[col]
    df_li = df.values.tolist()
    result = []
    for s in df_li:
        result.append(s)
    return result
```

```python
x = excel_one_line_to_list(df, '会计事务所名称')
x
```

```python
df.replace(x[:4], [0]*4, inplace = True)
df.replace(x[4:10], [1]*len(x[4:10]), inplace = True)
df.replace(x[10:], [2]*len(x[10:]))
```
