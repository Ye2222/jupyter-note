```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'A': 'foo bar foo bar foo bar foo foo'.split(),
                   'B': 'one one two three two two one three'.split(),
                   'C': np.arange(8), 'D': np.arange(8) * 2})
df
```

```python
def excel_one_line_to_list(df, col):
    df = df[col]
    df_li = df.values.tolist()
    result = []
    for s_li in df_li:
        result.append(s_li)
    print(result)
    return result
```

```python
x = excel_one_line_to_list(df, 'A')
x
```
