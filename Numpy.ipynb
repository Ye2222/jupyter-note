```python
# 进行向量的相加
def numpysum(n):
    a = np.arange(n) ** 2
    b = np.arange(n) ** 2
    print(a)
    print(b)
    c = a + b
    return c

numpysum(2)
```

```python
# 组合数组
c = np.arange(9).reshape(3, 3)
c
```

```python
d = 2 * c
d
```

```python
# 水平组合，使用hstack函数
np.hstack((c, d))
```

```python
# concatenate函数来实现同样的效果
np.concatenate((c, d), axis = 1)
```

```python
# 垂直组合，使用vstack函数
np.vstack((c, d))
```

```python
# concatenate函数也可实现这样的效果
np.concatenate((c, d), axis = 0)
```

```python
#  深度组合
# 将相同的元祖作为参数传给dstack函数，即可完成数组的深度组合
# 深度组合：将一系列数组沿着纵轴（深度）方向进行层叠组合
np.dstack((c, d))
```

```python
# 列组合 
# column_stack函数对于一维数组按排列方向进行组合
# 二维数组与水平排列效果一致
oned = np.arange(2)
twice_oned = 2 * oned
np.column_stack((oned, twice_oned)), np.column_stack((c, d))
```

```python
# 行组合
# row_stack函数按行方向进行组合
# 对于一维数组，将直接层叠起来组合成一个二维数组
np.row_stack((oned, twice_oned))
```

```python
# 对于二维数组，与垂直排列效果一致
np.row_stack((c, d))
```
