```python
def backward(x, y_hat, y, h):
    global w1, w2
    grad_y_hat = y_hat - y
    grad_w2 = h.T.dot(grad_y_hat)
    grad_h = grad_y_hat.dot(w2.T)
    grad_w1 = x.T.dot(grad_h * h * (1 - h))
    w1 -= lr * grad_w1
    w2 -= lr * grad_w2
```
